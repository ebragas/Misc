Import requests module for making HTTP requests, BeautifulSoup for parsing responses, and csv for writing the output to disk

```python
import requests
import csv
from bs4 import BeautifulSoup
```

Constants

```python
URL = r'https://www.tesla.com/findus/list/superchargers/United+States'
```

Make HTTP request and parse results into soup

```python
r = requests.get(URL)
soup = BeautifulSoup(r.text, 'lxml')
r.close()
```

Find all the states sections in page

```python
states = soup.find_all('div', {'class': 'state'})
```

```python
len(states)
for i in states:
    print(i.h2.text)
```

For each state section, find all the address tags

```python
data = []
row = {}

for state in states:
    
    for card in state.find_all('address', {'class': 'vcard'}):
        row['state'] = state.h2.text
        row['name'] = card.a.text
        row['url'] = card.a['href']
        row['street-address'] = card.find('span', {'class': 'street-address'}).text
        row['extended-address'] = card.find('span', {'class': 'extended-address'}).text
        row['locality'] = card.find('span', {'class': 'locality'}).text
        row['availability'] = "Coming Soon" if "(coming soon)" in row['name'] else "Available"
#         print(row)
        data.append(row.copy())
#         print(data[-1])
        
print(data)
print(len(data))
```

Write results to CSV file

```python
fieldnames = ['state', 'name','url', 'street-address', 'extended-address', 'locality', 'availability']

with open('c:\data\superchargers.csv', mode='w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    writer.writeheader()
    for r in data:
        writer.writerow(r)
```
